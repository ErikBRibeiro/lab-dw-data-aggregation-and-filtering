```python
import pandas as pd
import numpy as np
import cleaning

pd.options.display.max_columns = None

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'

df = pd.read_csv(url)

df.head()
```

```python
df = cleaning.standardize_column_names(df)

columns = ['unnamed:_0', 'customer']

df = cleaning.drop_columns(df, columns)
```

```python
df.info()
```

```python
new_df = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]

new_df
```

```python
average_claim_by_policy_gender = (df[df['response'] == 'Yes']).groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
average_claim_by_policy_gender
```

```python
customers_by_state = df.groupby('state').size()

states_over_500_customers = customers_by_state[customers_by_state > 500]

states_over_500_customers
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
df['month'] = pd.to_datetime(df['effective_to_date']).dt.month

policies_by_state_month = df.groupby(['state', 'month']).size().reset_index(name='number_of_policies')

total_policies_by_state = policies_by_state_month.groupby('state')['number_of_policies'].sum()

top_3_states = total_policies_by_state.nlargest(3).index

top_3_states_data = policies_by_state_month[policies_by_state_month['state'].isin(top_3_states)]

top_3_states_data
```
